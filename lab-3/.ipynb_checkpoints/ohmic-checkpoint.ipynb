Begin by importing the usual libraries.

```python
# This controls how plots look and behave in the Notebook
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

plt.rcParams['figure.figsize'] = [10, 6] # This command rests the default plot size to something bigger.
```

In the next cell enter your data for the Resistor, Lamp and LED.

```python
# Here we setup an array for the voltage and current data for each of the three components.
# Just enter your values separated by commas.
resistor_v = np.array([0.0, 1.0])
resistor_i = np.array([0.0, 1.0])
lamp_v = np.array([0.0, 1.0])
lamp_i = np.array([0.0, 1.0])
led_v = np.array([0.0, 1.0])
led_i = np.array([0.0, 1.0])
```

Plot the resistor data to see what it looks like.

```python
fig = plt.figure() # Create a plot figure called 'fig'.
ax = fig.add_subplot(111) # Define a set of axes called 'ax'.

ax.plot(resistor_i,resistor_v, 'ok') # Plot the data in resistor_i as the x-axis variable and
                                     # the data in resistor_v as the y-axis variable.
                                     # Use circles for data points "o" colored black "k".

# Set the title and axes labels for the plot.
ax.set_title('Your Title Here')
ax.set_xlabel('Your X-Axis Label Here')
ax.set_ylabel('Your Y-Axis Label Here')
```

Now fit the resistor data to a straight line and plot the fit along with the data.

```python
fit_param = np.polyfit(resistor_i, resistor_v,1) # This function fits the data to a polynomial of order n
                               # which in this case we set to n=1 which is a linear function.
                               # The best fit values for the slope and y-intercept are returned to
                               # the array fit_param

# Print out the best fit parameters for the slope and y-interecept.
print('Best fit Slope = ', fit_param[0], ' Best fit Y-intercept = ', fit_param[1])

resistor_fit_v = fit_param[1] + fit_param[0] * resistor_i # Use the best fit parameters to generate voltages for the best fit line.

# Create a new plot.
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(resistor_i, resistor_v, 'ok') # Plot the data as before.
ax.plot(resistor_i, resistor_fit_v, '-r') # Plot the best fit as a line "-" colored red "r".

ax.set_title('Your Title Here')
ax.set_xlabel('Your X-Axis Label Here')
ax.set_ylabel('Your Y-Axis Label Here')
```

Plot the Lamp data to see what it looks like.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(lamp_i, lamp_v, 'ok')

ax.set_title('Your Title Here')
ax.set_xlabel('Your X-Axis Label Here')
ax.set_ylabel('Your Y-Axis Label Here')
```

```python
fit_param = np.polyfit(lamp_i, lamp_v,1) # This function fits the data to a polynomial of order n
                               # which in this case we set to n=1 which is a linear function.
                               # The best fit values for the slope and y-intercept are returned to
                               # the array fit_param

print('Best fit Slope = ', fit_param[0], ' Best fit Y-intercept = ', fit_param[1])
lamp_fit_v = fit_param[1] + fit_param[0] * lamp_i # Use the best fit parameters to generate voltages for the best fit line.

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(lamp_i, lamp_v, 'ok')
ax.plot(lamp_i, lamp_fit_v, '-r')
ax.set_title('Your Title Here')
ax.set_xlabel('Your X-Axis Label Here')
ax.set_ylabel('Your Y-Axis Label Here')
```
